## Introducción

Zyfra es una empresa que desarrolla soluciones de eficiencia y seguridad para las industrias minera, petrolera, química y de la ingeniería. Una de las formas en las que se puede ayudar al mejoramiento de las industrias pesadas, es a través de la implementación de soluciones digitales, que permitan la optimización de la producción y eliminen parámetros no rentables. Es así que Zyfra requiere de la creación de un prototipo de modelo de machine learning que ayude a predecir la cantidad de oro extraído del mineral de oro.

Dentro del proceso de extracción de oro, el mineral extraído se somete a un tratamiento primario para obtener la materia prima o mezcla mineral que atravesará los procesos de flotación y purificación. Durante la flotación o proceso rougher, la mezcla mineral de oro se introduce en plantas de flotación, obteniendo dos productos importantes: el concentrado rougher (oro) y la cola rougher (residuos del producto). Posteriormente, el concentrado de oro se somete a dos etapas de purificación, a partir de la cual se obtendrá el concentrado final de oro y nuevas colas de residuos. 

Considerando esto, se requerirá construir un modelo que pueda predecir los valores del concentrado de oro rougher y el concentrado final. Para lo cual se contará con la información sobre los residuos y concentrados producidos en cada etapa de extracción y sus diferentes parámetros. 

### Objetivos

1. Desarrollar un modelo predictivo que permita establecer la cantidad de oro extraído a partir del mineral oro, considerando los residuos y concentrados que se producen en cada etapa del proceso. 
2. Realizar el preprocesamiento de datos y comprobar que los cálculos de recuperación sean correctos.
3. Analizar los datos a través de la exploración de distribuciones de metales y el tamaño de las particulas de alimentación. 
4. Entrenar diferentes modelos y evaluarlos a través de la métrica sMAPE (error medio absoluto porcentual simétrico) y la validación cruzada.

### Cálculo Recuperación

Una vez realizada la exploración inicial de nuestros tres datasets, procedemos a verificar si los valores de recuperación son los correctos. Para esto vamos a calcular la recuperación de la característica `rougher.output.recovery` del conjunto de entrenamiento, y estableceremos el EAM entre nuestros cálculos y los valores de la característica mencionada. 

Para el cálculo de la recuperación utilizaremos la siguiente fórmula:

$$ Recuperación =  \frac{C*(F - T)}{F*(C - T)}\,*100\% $$

dónde:
- C — proporción de oro en el concentrado justo después de la flotación (para saber la recuperación del concentrado rougher)/después de la purificación (para saber la recuperación del concentrado final)
- F — la proporción de oro en la alimentación antes de la flotación (para saber la recuperación del concentrado rougher)/en el concentrado justo después de la flotación (para saber la recuperación del concentrado final)
- T — la proporción de oro en las colas rougher justo después de la flotación (para saber la recuperación del concentrado rougher)/después de la purificación (para saber la recuperación del concentrado final)

Para calcular la recuperación vamos a crear la función `recovery_calculation` que tomará como argumentos el nombre del dataset y la etapa a partir de la cual se calculará la recuperación. En este caso para calcular la recuperación del concentrado Rougher se tomará en cuenta las características `rougher.output.concentrate_au`, `rougher.input.feed_au` y `'rougher.output.tail_au`; mientras que para calcular el concentrado final se considerará las caracaterísticas `final.output.concentrate_au`,  `rougher.output.concentrate_au` y `final.output.tail_au`. A partir de la cuales se calculará la recuperación en base a la fórmula antes presentada. 

```python
def recovery_calculation(df, stage=''):
    
    """
    df: dataframe base con el cual se realizará el cálculo
    stage: etapa utilizada para realizar el cálculo

    La función devolverá la siguiente operación, dependiendo de la etapa seleccionada:

    (C*(F-T))/(F*(C-T))*100
    """
    
    if stage == 'rougher':
        C = df['rougher.output.concentrate_au']
        F = df['rougher.input.feed_au']
        T = df['rougher.output.tail_au']
        
    if stage == 'final':
        C = df['final.output.concentrate_au']
        F = df['rougher.output.concentrate_au']
        T = df['final.output.tail_au']
        
    recovery = (C*(F-T))/(F*(C-T))*100
    
    return recovery
    
```

Establecida nuestra función `recovery_calculation`, vamos a filtrar nuestro dataset de entrenamiento con las características necesarias para el cálculo de recuperación del concentrado Rougher, lo guardaremos como `rougher_dataset` y eliminaremos los valores ausentes para que no ocasionen problemas posteriores en el cálculo del EAM. 

```python
rougher_dataset = df_train[['rougher.output.concentrate_au',
                             'rougher.input.feed_au',
                             'rougher.output.tail_au', 
                             'rougher.output.recovery']].dropna()

rougher_dataset.info()
```

A continuación vamos a aplicar la función `recovery_calculation` a nuestro nuevo dataset `roughet_dataset`, y compararemos nuestros cálculos de recuperación y los valores de la característica `rougher.output.recovery` a través del Error Absoluto Medio (EAM), para esto llamaremos a la función `mean_absolute_error` del módulo metrics de la librería Scikit-Learn. 

```python
rougher_calculated = recovery_calculation(rougher_dataset, 'rougher')
rougher_calculated.head()
```

```python
rougher_original = rougher_dataset['rougher.output.recovery']
rougher_original.head()
```

```python
print('EAM Recuperación Concentrado Rougher:', mean_absolute_error(rougher_calculated, rougher_original))
```

Los cinco primeros valores tanto de la recuperación calculada como los valores originales son idénticos, esto lo podemos corroborar con el valor del EAM  cercano a cero, lo que nos permite establecer que, a pesar de que existe un cierto grado de error, los valores de recuperación de nuestro conjunto de datos para el entrenamiento son similares a los valores calculados a partir de la fórmula de recuperación, así que podremos utilizar los valores originales para construir nuestros modelos predictivos.

### Cálculo valor final de sMAPE

Para evaluar nuestros modelos utilizaremos la métrica sMAPE o error medio absoluto porcentual simétrico. Es similar al MAE, pero se expresa en valores relativos en lugar de absolutos, y es simétrico porque tiene en cuenta la escala tanto del objetivo como de la predicción. Así se calcula:

$$ sMAPE =  \frac{1}{N}\sum\limits_{i=1}^{N}\frac{|y_{i}-\hat{y}_{i}|}{(|y_{i}|+|\hat{y}_{i}|) / 2}*100\%\ $$

- $y_{i}$ : valor del objetivo para la observación con el índice i en el conjunto utilizado para medir la calidad.
- $\hat{y}_{i}$ : Valor de la predicción para la observación con el índice i, por ejemplo, en la muestra de prueba.
- $N$ : número de observaciones de la muestra.
- $\sum\limits_{i=1}^{N}$: suma de todas las observaciones de la muestra (i toma valores de 1 a N)

La métrica final sMAPE incluye los dos valores que estamos tratando de predecir la recuperación del concentrado rougher `rougher.output.recovery` y la recuperación del concentrado final `final.output.recovery`:

$$ sMAPE final =  25\% * sMAPE(rougher) + 75\% * sMAPE(final) $$

Para calcular el valor final de sMAPE vamos a construir una función que tomará como argumentos los valores del objetivo y predicciones tanto del concentrado rougher como del concentrado final, y determinará la métrica final en base a las fórmulas previamente mencionadas. Nuestra función se llamará `smape_final` y tomará como argumentos:

- `y`: valor del objetivo, en este caso se trabajará con dos objetivos.
- `y_pred`: valor de las predicciones, se generarán dos predicciones.

```python
def smape_final(y, y_pred, **kwargs):
    # Definimos el número de observaciones de la muestra
    n = len(y)
    
    # Guardamos los valores absolutos del objetivo, predicciones y de su diferencia para el concentrado rougher
    y_rougher = abs(y.iloc[:, 0])
    y_pred_rougher = abs(y_pred[:, 0])
    diff_rougher = abs(y.iloc[:, 0] - y_pred[:, 0])
    
    # Guardamos los valores absolutos del objetivo, predicciones y de su diferencia para el concentrado final
    y_final = abs(y.iloc[:, 1])
    y_pred_final = abs(y_pred[:, 1])
    diff_final = abs(y.iloc[:, 1] - y_pred[:, 1])
    
    # Calculamos el smape para concentrado rougher y final 
    smape_rougher = (1/n) * np.sum(diff_rougher / ((y_rougher + y_pred_rougher)/2)) * 100
    
    smape_final = (1/n) * np.sum(diff_final / ((y_final + y_pred_final)/2)) * 100
    
    # Calculamos el smape final 
    smape_final_score = (0.25*smape_rougher) + (0.75*smape_final)

    return smape_final_score
```

Una vez establecida nuestra función para la métrica final de sMAPE, vamos a transformarla en un scorer o puntuador para que pueda ser posteriormente utilizada dentro de la validación cruzada. Para esto llamaremos a la función `make_scorer` de la librería Scikit-learn y pasaremos como argumento nuestra función `smape_final`. Guardaremos nuestro scorer como `smape_scorer`

```python
smape_scorer = make_scorer(smape_final)
```

```python
# Definimos nuestro modelo de regresión lineal
lr_model = LinearRegression()

# Obtenemos los valores de sMAPE para cada iteración de validación cruzada
lr_scores = cross_val_score(lr_model, features_train, target_train, scoring=smape_scorer, cv=5)

# Obtenemos el promedio de los cinco puntajes 
lr_final_score = lr_scores.mean()

print('Puntajes sMAPE para cada iteración:', lr_scores)
print('Modelo de Regresión Lineal | sMAPE = {:.6f}'.format(lr_final_score))
```

**Modelo de Árbol de Decisión**

```python
# Establecemos contadores para guardar el mejor puntaje y parámetros del modelo
best_score = float("inf")
best_depth = 0
best_split = 0
best_leaf = 0

# Generamos bucles que iteren sobre diferentes valores de max_depth, min_samples_split y min_sample_leaf
for depth in range(1, 6):
    
    for split in range(2, 9, 2):
        
        for leaf in range(2, 9, 2):
            # Definimos nuestro modelo y pasamos el conjunto de entrenamiento por validación cruzada
            dt_model = DecisionTreeRegressor(random_state=12345, max_depth=depth, min_samples_split=split, min_samples_leaf=leaf)
            dt_scores = cross_val_score(dt_model, features_train, target_train, scoring=smape_scorer, cv=5)
            
            # Obtenemos el promedio de los puntajes en cada iteración
            dt_final_score = dt_scores.mean()
            
            # Establecemos un bloque if para obtener los mejores hiperparámetros para el menor puntaje
            if dt_final_score < best_score:
                
                best_score = dt_final_score
                best_depth = depth
                best_split = split
                best_leaf = leaf

print(
    
    "Modelo Árbol de Decisión (max_depth = {}, min_samples_split = {}, min_samples_leaf = {}) | sMAPE = {:.6f}".format(
        best_depth, best_split, best_leaf, best_score)
)
```

**Modelo de Bosque Aleatorio**

```python
# Establecemos contadores para guardar el mejor puntaje y parámetros del modelo
best_score = float("inf")
best_depth = 0
best_est = 0

# Generamos bucles que iteren sobre diferentes valores de max_depth y n_estimators
for est in range(10, 41, 10):
    
    for depth in range(1, 5):
        # Definimos nuestro modelo y pasamos el conjunto de entrenamiento por validación cruzada
        rf_model = RandomForestRegressor(random_state=12345, max_depth=depth, n_estimators=est)
        rf_scores = cross_val_score(rf_model, features_train, target_train, scoring=smape_scorer, cv=5)
        
        # Obtenemos el promedio de los puntajes en cada iteración
        rf_final_score = rf_scores.mean()
        
        # Establecemos un bloque if para obtener los mejores hiperparámetros para el menor puntaje
        if rf_final_score < best_score:
                
                best_score = rf_final_score
                best_depth = depth
                best_est = est

print(
    
    "Modelo Bosque Aleatorio (max_depth = {}, n_estimators = {}) | sMAPE = {:.6f}".format(
        best_depth, best_est, best_score)
)
```

El modelo con el sMAPE más bajo 9.2854 corresponde al modelo de Bosque Aleatorio para una profundidad máxima de 4 y un número de estimadores de 40. El Árbol de Decisión se encuentra en segundo lugar, con un sMAPE de 10.0506 y los mejores hiperparámetros fueron 2 de profundidad máxima, 2 para el número mínimo de división de muestras, y 2 para el número mínimo de hojas. El modelo de regresión lineal fue el que obtuvo el puntaje sMAPE más alto de 10.7564.

En base a estos resultados, decidimos quedarnos con el modelo de Bosque Aleatorio como el mejor modelo para predecir el concentrado rougher y concentrado final de oro, ya que obtuvo el sMAPE más bajo.
